```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def function_value(x1, x2, Y):
  return (x1**2 + Y*(x2**2) -x1*x2 - x1 - x2)
```

```python
#calculate gradients using partial diffrentiation
def dx1(x1, x2):
  return (2*x1 - x2 - 1)
def dx2(x1, x2, Y):
  return (2*Y*x2 - x1 - 1)
```

Q2.1

```python
def gradient_descend(x1, x2, Y, learning_rate, epochs):

  n_iters = 0
  #min values of x1, x2 and function being stored in these variables
  min_x1 = x1
  min_x2 = x2
  y_pred = sys.maxsize
  y_pred_array = np.array([])

  for n_iters in range(epochs):
    min_x1 = x1 - learning_rate*dx1(x1, x2)  #wt updation
    min_x2 = x2 - learning_rate*dx2(x1, x2, Y)
    min_y_pred = function_value(min_x1, min_x2, Y)

    if (y_pred - min_y_pred) < convergence_tolerance :
      print('\n')
      break
    else :
      x1 = min_x1
      x2 = min_x2
      y_pred = min_y_pred
      print(f'epoch:{n_iters + 1}  x1:{x1}  x2:{x2}  function_value:{y_pred}')
      y_pred_array = np.append(y_pred_array , y_pred)
  
  return (n_iters), min_x1, min_x2, min_y_pred, y_pred_array
   
```

```python
#hyperparameters
learning_rate = 0.1
epochs = 1000
Y = 1
convergence_tolerance = 10**-4

#taking inputs as x1 = 5, x2 = 2; executing the algorithm
n_iters, min_x1, min_x2, min_y_pred, y_pred_array = gradient_descend(5,2,Y, learning_rate, epochs)  
# print(n_iters, min_x1, min_x2, min_y_pred)
```
